```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("casasboston.csv")

print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]

df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})

print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist()
plt.show()
```

Correccion

```python
df['N_HABITACIONES_MEDIO'].plot.hist(color='skyblue', edgecolor='black')
plt.title('Histograma del Número Medio de Habitaciones')
plt.xlabel('Número Medio de Habitaciones')
plt.ylabel('Frecuencia')
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1)
plt.show()
```

Correccion

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN", alpha=0.5,color='skyblue')

plt.title('Relación entre el Valor Mediano y el Índice de Crimen')
plt.xlabel('Valor Mediano')
plt.ylabel('Índice de Crimen')

plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head(10).plot.barh()
plt.show()
```

Correccion

```python
valor_por_ciudad = df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head(10).plot.barh(color='skyblue',edgecolor='black')

plt.title('Valor Mediano de las Viviendas por Ciudad (Top 10)')
plt.xlabel('Valor Mediano')
plt.ylabel('Ciudad')

plt.show()
```
